```python
import serial
import time

# Substitua pela porta COM correta do seu ESP32
PORTA = "COM7"
BAUD = 115200  # ou 115200, conforme o código do ESP32

try:
    print(f"Conectando à porta {PORTA}...")
    esp32 = serial.Serial(PORTA, BAUD, timeout=2)
    time.sleep(2)  # aguarda a conexão estabilizar
    print("Conectado com sucesso!\nLendo dados:\n")

    while True:
        if esp32.in_waiting > 0:  # se há dados disponíveis
            dado = esp32.readline().decode(errors='ignore').strip()
            print(f"Recebido: {dado}")

except serial.SerialException:
    print(f"❌ Erro: não foi possível abrir a porta {PORTA}. Verifique o nome e se está livre.")
except KeyboardInterrupt:
    print("\nPrograma encerrado.")
finally:
    if 'esp32' in locals() and esp32.is_open:
        esp32.close()
```
